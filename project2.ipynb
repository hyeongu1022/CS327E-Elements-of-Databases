## INNER JOIN

#### Using INNER JOIN to retrieve a table with actors' information and their rated-PG movies, ordered by their film_id. 

```python
!psql -c "select actor.actor_id, actor.first_name, actor.last_name, film.film_id, film.title, film.rating \
            FROM pagila.actor INNER JOIN pagila.film_actor ON actor.actor_id = film_actor.actor_id \
            INNER JOIN pagila.film ON film_actor.film_id = film.film_id \
            WHERE film.rating='PG' ORDER BY film_id LIMIT 10"
```

#### Using INNER JOIN to retrieve English movies that have comparably low rental rate (less than 3) with their description ordered by their rental rate.

```python
!psql -c "SELECT film.film_id, film.title, film.description, film.language_id, language.name, film.rental_rate \
            FROM pagila.film INNER JOIN pagila.language ON film.language_id = language.language_id \
            WHERE film.rental_rate < 3 ORDER BY rental_rate LIMIT 10"
```

#### Using INNER JOIN to retrive data table of each city's id, city, country id and we used INNER JOIN with city table and address table on city_id

```python
!psql -c "select city.city_id, city.city, city.country_id, address.city_id,\
    address.address_id, address.postal_code \
    FROM pagila.city INNER JOIN pagila.address ON city.city_id = address.city_id \
    WHERE city.city_id < '10' ORDER BY city LIMIT 10"
```

## OUTER JOIN

#### Using LEFT JOIN to create a table with addresses of stores in Mexico ordered by their addresses. 

```python
!psql -c "SELECT country.country_id, country.country, city.city, address.address, address.postal_code \
            FROM pagila.country LEFT JOIN pagila.city ON country.country_id = city.country_id \
            LEFT JOIN pagila.address ON address.city_id = city.city_id \
            WHERE country.country = 'Mexico' ORDER BY address LIMIT 10"
```

#### Using RIGHT JOIN to see the movies with language English.

```python
!psql -c "SELECT film.film_id, film.title, film.language_id, language.name \
            FROM pagila.film RIGHT JOIN pagila.language ON film.language_id = language.language_id \
            WHERE language.language_id = 1 ORDER BY title LIMIT 10"
```

#### We used FULL OUTER JOIN to retirve data table each cutomer's store id, customer id, create date, and we used OUTER JOIN with cutomer table and store table on customer_id


```python
!psql -c "select customer.store_id, customer.customer_id, customer.create_date, store.store_id, \
store.manager_staff_id, store.address_id FROM pagila.customer FULL OUTER JOIN pagila.store \
ON customer.store_id < store.store_id WHERE customer_id >'5' ORDER by address_id LIMIT 10"
```
